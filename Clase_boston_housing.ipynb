```python
df_limpio = df.dropna()
df_limpio.describe()
```

```python
lm = LinearRegression()
x1 = df_limpio['CRIM'].values.reshape(-1,1)
x2 = df_limpio['ZN'].values.reshape(-1,1)
x3 = df_limpio['INDUS'].values.reshape(-1,1)
x4 = df_limpio['CHAS'].values.reshape(-1,1)
x5 = df_limpio['NOX'].values.reshape(-1,1)
x6 = df_limpio['RM'].values.reshape(-1,1)
x7 = df_limpio['AGE'].values.reshape(-1,1)
x8 = df_limpio['DIS'].values.reshape(-1,1)
x9 = df_limpio['RAD'].values.reshape(-1,1)
x10 = df_limpio['TAX'].values.reshape(-1,1)
x11 = df_limpio['PTRATIO'].values.reshape(-1,1)
x12 = df_limpio['B'].values.reshape(-1,1)
x13 = df_limpio['LSTAT'].values.reshape(-1,1)
y = df_limpio['MEDV']
```

```python
import statsmodels.api as sm
```

```python
n = len(df_limpio)
ones = np.ones([n,1])
x = np.hstack([ones, x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11, x12, x13])
```

```python
ols = sm.OLS(y, x)
ols_result = ols.fit()
ols_result.summary()
```

```python
df_noduplicados = df_limpio.drop_duplicates()
df_noduplicados.describe()
```

```python
for col in df_noduplicados.columns:
    if col == 'MEDV': break
    min_val = df_noduplicados[col].min()
    max_val = df_noduplicados[col].max()
    
    aux = (df_noduplicados - min_val) / (max_val - min_val)
    df_noduplicados[col] = aux[col]

medv = df_noduplicados['MEDV']
```
