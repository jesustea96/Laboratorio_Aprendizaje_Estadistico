```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
df = pd.read_excel('Motor Trend Car Road Tests.xlsx')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
y = df['mpg']
x1 = df['cyl'].values.reshape(-1, 1)
x2 = df['disp'].values.reshape(-1, 1)   
x3 = df['hp'].values.reshape(-1, 1)
x4 = df['drat'].values.reshape(-1, 1)
x5 = df['wt'].values.reshape(-1, 1)
x6 = df['qsec'].values.reshape(-1, 1)
x7 = df['vs'].values.reshape(-1, 1)
x8 = df['am'].values.reshape(-1, 1)
x9 = df['gear'].values.reshape(-1, 1)
x10 = df['carb'].values.reshape(-1, 1)
```

```python
n = len(df)
ones = np.ones([n,1])
x = np.hstack([ones, x1, x2, x3, x4, x5, x6, x7, x8, x9, x10])
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.4, random_state=137)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_xtrain = StandardScaler().fit(x_train)
x_train = scaler_xtrain.transform(x_train)
scaler_xtest = StandardScaler().fit(x_test)
x_test = scaler_xtest.transform(x_test)
```

```python
n = len(x_train)
ones = np.ones([n,1])
x = np.hstack([ones, x_train])
```

```python
ols = sm.OLS(y_train, x)
ols_result = ols.fit()
ols_result.summary()
```

```python
model_pred = ols_result.predict(x_test)
model_pred
```
